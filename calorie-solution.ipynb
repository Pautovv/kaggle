```python
%%time
import numpy as np, pandas as pd, xgboost as xgb, optuna
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
data_train = pd.read_csv('/kaggle/input/playground-series-s5e5/train.csv')
data_test = pd.read_csv('/kaggle/input/playground-series-s5e5/test.csv')
data_train.head(2)
```

```python
data_test.head(2)
```

```python
print(f"{data_train.shape}\n{data_test.shape}")
```

```python
for col in data_train.columns.to_list():
    print(f"{col} : {data_train[col].isna().sum()}")
```

```python
for col in data_test.columns.to_list():
    print(f"{col} : {data_test[col].isna().sum()}")
```

```python
data_train['Sex'] = LabelEncoder().fit_transform(data_train['Sex'])
data_test['Sex'] = LabelEncoder().fit_transform(data_test['Sex'])
print('done')
```

```python
X_train, y_train  = data_train.iloc[:, :-1], data_train.iloc[:, -1]
print('done')
```

```python
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
print('done')
```

```python
#boost = xgb.XGBoostRegressor()
#boost.fit(X_train, y_train)
```

```python
y_pred = forest.predict(data_test)
print('done')
```

```python
submission = pd.DataFrame({
    'id': data_test['id'],
    'Calories': y_pred
})
submission.to_csv('submission.csv', index=False)
print('done')
```
